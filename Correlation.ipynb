```python
import pandas as pd
```

```python
data=pd.read_csv('Ames_Housing_Data1.tsv',sep='\t')
```

```python
print(data)
```

```python
#select only numeric data
N_data=data.select_dtypes(include=['int64','float64'])
print(N_data)
```

```python
N_data_corr=N_data.corr()['SalePrice'][:-1]
print(N_data_corr)
```

```python
#strong correlatiom >0.5
Top_features=N_data_corr[abs(N_data_corr)>0.5].sort_values(ascending=False)
print(Top_features)
```
